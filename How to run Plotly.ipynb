# Plotly : Interactive visualization tool

Used for visualize diffrent insights of data.

```python
# To run/visualize graphs through plotly:
1. Open Google colab >> Create New Notebook
2. import plotly.graph_objects as go
3. Use different methods to plot graphs similar to matplotlib or Seaborn.
```

```python
# for example:
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[3,4,5,6,7], mode= 'markers'))
fig.show() 

'''
Explanation: 

1. 'fig refers' to a variable that represents a plotly figure object, which can contain one or more traces.
2. 'add_trace' is a method of the figure object that allows you to add one or more traces to the figure.
3. 'go.Scatter' is a function from the plotly.graph_objs module that creates a scatter trace.
4. The x argument is a list or array of x-coordinates for the scatter plot.
5. The y argument is a list or array of y-coordinates for the scatter plot.
6. The 'mode' argument specifies the display mode of the scatter plot, 
   and 'markers' indicates that the plot will display only the marker points 
   and not any lines or other connecting elements.

Overall, this code adds a scatter trace to the plotly figure object fig, with x-coordinates [1,2,3,4,5] and y-coordinates [3,4,5,6,7], 
and with the display mode set to 'markers'.
'''

# On hovering over points, we can visualize (x,y) values
# Graphs can be easily downloaded.
```
