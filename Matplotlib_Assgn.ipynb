```python
# Ans 1:
'''
Matplotlib is a popular plotting library for Python that allows us to create a wide range of 2D and 3D plots and visualizations.
It is an open-source library that can be used in various Python environments, including Jupyter notebooks, web applications, and standalone Python scripts.
'''
'''
It's uses:
1. Matplotlib is widely used in the data science community for data visualization and analysis tasks.
2. It offers a large range of customization options, making it suitable for creating high-quality, publication-ready visualizations.
3. It is compatible with various operating systems and offers a range of output formats, including PNG, PDF, SVG, and more.
4. It can be used to plot a variety of data, including time-series data, statistical data, and geospatial data.
'''
'''
Five plots that can be plotted using the Pyplot module of Matplotlib:

>> Line plot: Used to plot continuous data over time or other numerical domains.
>> Scatter plot: Used to visualize the relationship between two continuous variables.
>> Bar plot: Used to show comparisons between different categories of data.
>> Histogram: Used to show the distribution of a continuous variable.
>> Box plot: Used to show the distribution of a dataset and identify any outliers.
'''
```

```python
# Ans 2:
'''
>> A scatter plot shows the relationship between two continuous variables.
>> It plots each data point as a point on a 2-dimensional coordinate system.
>> One variable is plotted on the x-axis, and the other is plotted on the y-axis.
>> They are useful for identifying patterns and trends in data.
>> They can be used to investigate the correlation between two variables.
>> Outliers and clusters can be identified using scatter plots.
>> They can be created using software tools such as Matplotlib and Excel.
'''
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y, s=100, c= 'g')
plt.title("Scatter plot", c = 'r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Ans 3:
'''
>> The subplot() function in Matplotlib is used to create multiple plots or subplots within a single figure.
>> It allows you to organize and visualize multiple plots in a single layout.
>> This is useful for comparing different data sets or displaying different aspects of the same data set.
>> The function allows you to customize the layout of the plots by specifying the number of rows and columns in the grid, 
   as well as the position of each individual subplot within the grid.
'''
import numpy as np
import matplotlib.pyplot as plt

# Create the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot each line on a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title for the entire plot
fig.suptitle('Line Plots')

# Display the plot
plt.show()
```

```python
# Ans 4:
'''
>> A bar plot is a type of graph used to compare different categories of data.

1. It displays data as rectangular bars with heights or lengths proportional to the values they represent.
2. The bars can be either vertical or horizontal, depending on the orientation of the graph.
3. Bar plots are used to visualize and compare the size or magnitude of different categories of data.
4. They are commonly used to display data that is discrete or categorical in nature.
5. Bar plots are useful for identifying trends, patterns, and outliers in the data.
6. They can be used to compare data across different time periods, different groups, or different categories.
7. Bar plots can be created using various software tools, including Matplotlib, Excel, and other data analysis software.
8. They are commonly used in business, finance, economics, and other fields where comparisons of discrete data are important.
'''
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.title("Vertical Bar plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
# Creating Horizontal Bar-plot:
plt.barh(company,profit)
plt.title("Horizontal Bar plot")
plt.ylabel("Company")
plt.xlabel("Profit")
plt.show()
```

```python
# Ans 5:
'''
Box Plot : It is a type of graph used to display the distribution of a dataset. 
           It is also known as a box-and-whisker plot. 
           The plot is based on a five-number summary of the data: the minimum value, the first quartile (Q1), the median, the third quartile (Q3), and the maximum value. 
           The box in the plot represents the interquartile range (IQR), which is the distance between the first and third quartiles. 
           The whiskers extend from the box to the minimum and maximum values, and any data points that lie beyond the whiskers are considered outliers.

Box plots are used to:

Identify the central tendency, spread, and skewness of the dataset.
Identify outliers in the dataset.
Compare the distributions of two or more datasets.
'''

import matplotlib.pyplot as plt
import numpy as np

# Create the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create the plot
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box plot of Box 1 and Box 2')
plt.show()
```
