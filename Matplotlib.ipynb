```python
# To add labels in x,y axis:
plt.plot(x,y)
plt.xlabel("This is my x-data")
plt.ylabel("This is my y-data")
plt.title("This is my Graph title")
plt.show()
```

# Some more examples:

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)  #To print scattered dot wise data
# Scattered plot data doesn't show any relationship but used for visualization.
```

```python
# To print dots in diff colors:
colours = np.random.rand(50)
plt.scatter(x,y, c=colours)
```

```python
# Even sizes of dots can be controlled:
sizes = np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes)  #Size is visually small
```

```python
# To view enlarged size of dots:
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes)
```

```python
# To view intensity/depth of dot colors:
sizes = 1000*np.random.rand(50)
# plt.scatter(x,y, c = colours, s = sizes, alpha=.5)
# plt.scatter(x,y, c = colours, s = sizes, alpha=.8)
plt.scatter(x,y, c = colours, s = sizes, alpha=.1) #Fade color
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```

# Bar - Graph: To display Categorical v/s Numerical data in a graph

```python
# Strings/ catergorical data:
x = ['a','b','c','d','e']
```

```python
# Random numerical values:
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Bar Plot - Vertical")
```

```python
y
```

```python
# To display bar-graph horizontally with same data: barh() method is used
# Here x-axis switched to y-axis and vice-versa
plt.barh(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Bar Plot - Horizontal")
```

# Using same sin-wave graph to perform more graph beautifications:


```python
# Creating some data for x-y axis
x = np.linspace(0,10 , 200)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("This is my x-data")
plt.ylabel("This is my y-data")
plt.title("This is my Graph title")
plt.show()
```

```python
# To create graph lines in diff color and not solid line
# plt.plot(x,y,'--r') #red line
plt.plot(x,y,'--b') #blue line
plt.xlabel("This is my x-data")
plt.ylabel("This is my y-data")
plt.title("This is my Graph title")
plt.show()
```

```python
# To control figure size: zoomed-in , zoomed-out
plt.figure(figsize=(8,4))
plt.plot(x,y,'--g')
plt.xlabel("This is my x-data")
plt.ylabel("This is my y-data")
plt.title("This is my Graph title")
plt.show()
```

```python
# More smaller size graph -window
plt.figure(figsize=(6,2))
plt.plot(x,y,'--r')
plt.xlabel("This is my x-data")
plt.ylabel("This is my y-data")
plt.title("This is my Graph title")
plt.show()
```

```python
# To display grid-structure in background:
plt.figure(figsize=(6,2))
plt.plot(x,y,'--y')
plt.xlabel("This is my x-data")
plt.ylabel("This is my y-data")
plt.title("This is my Graph title")
plt.grid()
plt.show()
```

# Using customization with Scattered plot() graph

```python
# Existing graph:
x= np.random.rand(50)
y= np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes, alpha=.5) #Fade color
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```

```python
# To define any particular color for dots: ex:green
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = 'g', s = sizes, alpha=.5) 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```

```python
# Instead of c = 'g' >> Any hexadecimal color code combination can be given from HTML/CSS
# for example:
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = '#FFA500', s = sizes, alpha=.5) 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```

```python
# To represent scattered data in diff shape instead of circular-dots:
# use marker = 'pattern to display'  only specific chars can be represented
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = '#FFA500', s = sizes, alpha=.5, marker='x') 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```

```python
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = '#FFA500', s = sizes, alpha=.5, marker='v') 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```

# 3-D Graphs --> Data having:[x-axis, y-axis, z-axis]

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
# To display data in 3-D plane:
fig = plt.figure() #creating figure
ax = fig.add_subplot(projection = '3d') # need to set axis, define projection of the image
ax.scatter(x,y,z)
plt.show()
```
